```python
#11
def is_valid_regex(s):
    n = len(s)
    i = 0
    prev_repeat = False  # đánh dấu ký tự trước có phải * hoặc +

    while i < n:
        if s[i] == '\\':
            # escape phải có ký tự theo sau
            if i + 1 >= n:
                return False
            prev_repeat = False
            i += 2
        elif s[i] in ['*', '+']:
            # không được đứng đầu hoặc lặp toán tử
            if i == 0 or prev_repeat:
                return False
            prev_repeat = True
            i += 1
        else:
            prev_repeat = False
            i += 1

    return True


# Đọc input
t = int(input())
for _ in range(t):
    s = input().strip()
    try:
        if is_valid_regex(s) == False :
            raise Exception("Invalid")
        else:
            print("Valid")
    except Exception as e:
        print(e)
```
