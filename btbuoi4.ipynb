```python
def perfect_squares_distinct(n: int):
    res = []
    i = 1
    while i * i <= n:
        sq = i * i
        s = str(sq)
        if len(set(s)) == len(s):
            res.append(s)
        i += 1
    return res
if __name__ == "__main__":
    try:
        n = int(input().strip())
        if n <= 0:
            print("no number")
        else:
            ans = perfect_squares_distinct(n)
            if ans:
                print(" ".join(ans))
            else:
                print("no number")
    except:
        print("no number")
```

```python
def collatz_length(n):
    length = 1
    while n != 1:
        if n % 2 == 0:
            n /= 2
        else:
            n = 3 * n + 1
        length += 1
    return length

n = int(input())
max_len = 0
num = 1
for i in range(1, n + 1):
    l = collatz_length(i)
    if l > max_len or (l == max_len and i < num):
        max_len = l
        num = i
print(num, max_len)
```

```python
def so_than_thiet(n):
    s = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            s += i
            if i != n // i:
                s += n // i
    return s

a, b = map(int, input().split())

if a != b and so_than_thiet(a) == b and so_than_thiet(b) == a:
    print("true")
else:
    print("false")
```

```python
# Nhập chuỗi từ bàn phím
chuoi = input("Nhập vào một chuỗi: ")

# Khởi tạo biến đếm
dem_in_hoa = 0
dem_in_thuong = 0
dem_so = 0

# Duyệt từng ký tự trong chuỗi
for ch in chuoi:
    if ch.isupper():        # kiểm tra chữ in hoa
        dem_in_hoa += 1
    elif ch.islower():      # kiểm tra chữ in thường
        dem_in_thuong += 1
    elif ch.isdigit():      # kiểm tra ký tự số
        dem_so += 1

# In kết quả
print("Số chữ in hoa:", dem_in_hoa)
print("Số chữ in thường:", dem_in_thuong)
print("Số ký tự số:", dem_so)
```

```python
def tinh_tong_cac_ky_tu_so(chuoi_dau_vao):
    # Khởi tạo tổng
    tong = 0

    # Chuỗi chứa các ký tự số được tìm thấy để minh họa
    cac_ky_tu_so = []

    # Lặp qua từng ký tự trong chuỗi đầu vào
    for ky_tu in chuoi_dau_vao:
        # Kiểm tra xem ký tự đó có phải là một chữ số ('0' đến '9') không
        if ky_tu.isdigit():
            # Chuyển ký tự số thành số nguyên (integer)
            gia_tri_so = int(ky_tu)

            # Cộng giá trị số vào tổng
            tong += gia_tri_so

            # Lưu lại ký tự số để hiển thị quá trình tính toán
            cac_ky_tu_so.append(ky_tu)

    # Xử lý chuỗi hiển thị tổng
    bieu_thuc_tong = " + ".join(cac_ky_tu_so)

    print(f"Các ký tự số được tìm thấy: {', '.join(cac_ky_tu_so)}")
    print(f"Biểu thức tổng: {bieu_thuc_tong}")

    return tong

# --- ÁP DỤNG VỚI VÍ DỤ CỤ THỂ ---
chuoi_mau = "abd45ecf47wde3s1"

print(f"Chuỗi nhập vào: {chuoi_mau}")
ket_qua = tinh_tong_cac_ky_tu_so(chuoi_mau)

print("---")
print(f"Tổng cuối cùng là: {ket_qua}")
```

```python
def la_mat_khau_manh(chuoi):
    co_hoa = co_thuong = co_so = co_dac_biet = False

    if len(chuoi) < 8:
        return False

    for ch in chuoi:
        if ch.isupper():
            co_hoa = True
        elif ch.islower():
            co_thuong = True
        elif ch.isdigit():
            co_so = True
        elif not ch.isalnum():
            co_dac_biet = True

    return co_hoa and co_thuong and co_so and co_dac_biet

# Nhập chuỗi
mk = input("Nhập mật khẩu: ")
if la_mat_khau_manh(mk):
    print("Đây là mật khẩu mạnh.")
else:
    print("Mật khẩu chưa đủ mạnh.")
```
