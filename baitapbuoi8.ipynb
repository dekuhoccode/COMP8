```python
#1
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        self.pi = 3.14
    def surface_area(self):
        return 2 * self.pi * self.radius**2 + 2 * self.pi * self.radius * self.height
    def volume(self):
        return self.pi * self.radius**2 * self.height
r, h = map(float, input().split())
cyl = Cylinder(r, h)
print(f"{cyl.surface_area():.2f} {cyl.volume():.2f}")
```

```python
#2
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    def is_valid(self):
        days_in_month = [31, 28, 31, 30, 31, 30,
                         31, 31, 30, 31, 30, 31]

        if (self.year % 400 == 0) or (self.year % 4 == 0 and self.year % 100 != 0):
            days_in_month[1] = 29

        if self.month < 1 or self.month > 12:
            return False
        if self.day < 1 or self.day > days_in_month[self.month - 1]:
            return False
        return True

    def next_day(self):
        days_in_month = [31, 28, 31, 30, 31, 30,
                         31, 31, 30, 31, 30, 31]
        if (self.year % 400 == 0) or (self.year % 4 == 0 and self.year % 100 != 0):
            days_in_month[1] = 29

        d, m, y = self.day, self.month, self.year

        d += 1
        if d > days_in_month[m - 1]:
            d = 1
            m += 1
            if m > 12:
                m = 1
                y += 1
        return Date(d, m, y)

date_str = input().strip()
try:
    dd, mm, yyyy = map(int, date_str.split("/"))
    d = Date(dd, mm, yyyy)
    if not d.is_valid():
        print("INVALID")
    else:
        next_d = d.next_day()
        print(f"{next_d.day:02d}/{next_d.month:02d}/{next_d.year:04d}")
except:
    print("INVALID")
```

```python
#4
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def set_numbers(self, a, b):
        self.a = a
        self.b = b

    def add(self):
        return self.a + self.b

    def subtract(self):
        return self.a - self.b

    def product(self):
        return self.a * self.b

    def divide(self):
        if self.b == 0:
            return "Error: Division by zero"
        return self.a / self.b

    def power(self):
        return self.a ** self.b

    def mod(self):
        if self.b == 0:
            return "Error: Modulo by zero"
        return self.a % self.b


# --- Chương trình chính ---
def main():
    a, b = map(float, input("Nhập 2 số ban đầu: ").split())
    calc = Calculator(a, b)

    while True:
        print("\n--- MENU ---")
        print("1. Cộng")
        print("2. Trừ")
        print("3. Nhân")
        print("4. Chia")
        print("5. Hàm mũ")
        print("6. Mod (phần dư)")
        print("7. Nhập lại 2 số mới")

        choice = input("Chọn phép toán (1-7): ")

        if choice == "1":
            print("Kết quả:", calc.add())
        elif choice == "2":
            print("Kết quả:", calc.subtract())
        elif choice == "3":
            print("Kết quả:", calc.product())
        elif choice == "4":
            print("Kết quả:", calc.divide())
        elif choice == "5":
            print("Kết quả:", calc.power())
        elif choice == "6":
            print("Kết quả:", calc.mod())
        elif choice == "7":
            a, b = map(float, input("Nhập 2 số mới: ").split())
            calc.set_numbers(a, b)
            print("Đã cập nhật 2 số mới.")
        else:
            print("Lựa chọn không hợp lệ!")

        exit_choice = input("Bạn có muốn thoát? (yes/YES/Yes để thoát): ")
        if exit_choice.lower() == "yes":
            print("Chương trình kết thúc.")
            break


if __name__ == "__main__":
    main()
```

```python
#7
import math

class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mẫu số không được bằng 0")
        self.num = num
        self.den = den
        self.simplify()

    def simplify(self):
        # Rút gọn phân số
        g = math.gcd(self.num, self.den)
        self.num //= g
        self.den //= g
        # Quy ước mẫu số dương
        if self.den < 0:
            self.num = -self.num
            self.den = -self.den

    def add(self, other):
        return Fraction(self.num * other.den + other.num * self.den,
                        self.den * other.den)

    def sub(self, other):
        return Fraction(self.num * other.den - other.num * self.den,
                        self.den * other.den)

    def mul(self, other):
        return Fraction(self.num * other.num, self.den * other.den)

    def div(self, other):
        return Fraction(self.num * other.den, self.den * other.num)

    def __str__(self):
        return f"{self.num}/{self.den}"


# ------------------ MAIN ------------------
def main():
    # Input dạng: a b op c d
    a, b, op, c, d = input().split()
    a, b, c, d = int(a), int(b), int(c), int(d)

    f1 = Fraction(a, b)
    f2 = Fraction(c, d)

    if op == '+':
        result = f1.add(f2)
    elif op == '-':
        result = f1.sub(f2)
    elif op == '*':
        result = f1.mul(f2)
    elif op == '/':
        result = f1.div(f2)
    else:
        raise ValueError("Phép toán không hợp lệ")

    print(result)


if __name__ == "__main__":
    main()
```

```python
#10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        return sum(1 for b in self.books if b.author == name)

    def find_by_year(self, y):
        return sum(1 for b in self.books if b.year == y)


# Xử lý input/output
def main():
    n = int(input().strip())
    library = Library()

    for _ in range(n):
        line = input().strip()
        if line.startswith("ADD"):
            # Ví dụ: ADD title; author,year
            _, rest = line.split(" ", 1)
            title_part, rest2 = rest.split(";")
            author_part, year_part = rest2.split(",")
            title = title_part.strip()
            author = author_part.strip()
            year = int(year_part.strip())
            library.add(Book(title, author, year))

        elif line.startswith("COUNTYEAR"):
            _, year_str = line.split(" ", 1)
            year = int(year_str.strip())
            print(library.find_by_year(year))

        elif line.startswith("COUNT"):
            _, author = line.split(" ", 1)
            print(library.count_by_author(author.strip()))


if __name__ == "__main__":
    main()
```

```python
#11
import math

def vector_sub(a, b):
    return [a[i] - b[i] for i in range(3)]

def cross(u, v):
    return [u[1]*v[2] - u[2]*v[1],
            u[2]*v[0] - u[0]*v[2],
            u[0]*v[1] - u[1]*v[0]]

def dot(u, v):
    return sum(u[i]*v[i] for i in range(3))

def norm(u):
    return math.sqrt(dot(u, u))

# Input
A = list(map(float, input().split()))
B = list(map(float, input().split()))
C = list(map(float, input().split()))
D = list(map(float, input().split()))

# Pháp tuyến mặt phẳng (A,B,C)
AB = vector_sub(B, A)
AC = vector_sub(C, A)
n1 = cross(AB, AC)

# Pháp tuyến mặt phẳng (B,C,D)
BC = vector_sub(C, B)
BD = vector_sub(D, B)
n2 = cross(BC, BD)

# Góc giữa hai mặt phẳng
cos_phi = dot(n1, n2) / (norm(n1) * norm(n2))
# tránh sai số vượt [-1,1]
cos_phi = max(-1, min(1, cos_phi))
phi = math.degrees(math.acos(cos_phi))

print(f"{phi:.2f}")
```

```python
#12
class ComplexNumber:
    def __init__(self, initRealPart, initImaginaryPart):
        self.realPart = initRealPart
        self.imaginaryPart = initImaginaryPart

    def multiply(self, other):
        real = self.realPart * other.realPart - self.imaginaryPart * other.imaginaryPart
        imag = self.realPart * other.imaginaryPart + self.imaginaryPart * other.realPart
        return ComplexNumber(real, imag)


def main():
    # Nhập số phức thứ nhất
    r1, i1 = map(float, input().split())
    c1 = ComplexNumber(r1, i1)

    # Nhập số phức thứ hai
    r2, i2 = map(float, input().split())
    c2 = ComplexNumber(r2, i2)

    # Nhân và in kết quả
    result = c1.multiply(c2)
    print(f"{int(result.realPart)} {int(result.imaginaryPart)}")


if __name__ == "__main__":
    main()
```

```python
#13
import math

class Triangle:
    def __init__(self, initEdge1, initEdge2, initEdge3):
        self.a = initEdge1
        self.b = initEdge2
        self.c = initEdge3

    def getArea(self):
        # Kiểm tra điều kiện tam giác
        if (self.a + self.b <= self.c or
            self.a + self.c <= self.b or
            self.b + self.c <= self.a or
            self.a <= 0 or self.b <= 0 or self.c <= 0):
            return -1  # không hợp lệ

        # Công thức Heron
        p = (self.a + self.b + self.c) / 2
        area = math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))
        return area


def main():
    # Nhập 3 cạnh
    edges = list(map(int, input().split()))
    t = Triangle(edges[0], edges[1], edges[2])
    area = t.getArea()

    if area == -11:
        print("invalid")
    else:
        print(f"{area:.2f}")


if __name__ == "__main__":
    main()
```
