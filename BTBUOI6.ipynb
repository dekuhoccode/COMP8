```python
#w6a4
s = input().split()

result = {}

for pair in s:
    key, value = pair.split(":")

    if key not in result:
        result[key] = []

    result[key].append(value)

print(result)
```

```python
#w6a9
import ast # Thư viện để chuyển chuỗi thành dict an toàn
# ast.literal_eval sẽ tự hiểu "{'a': 1}" là dictionary, và số 1 là int
text1 = ast.literal_eval(input())
text2 = ast.literal_eval(input())

result = text1.copy()
for key,value in text2.items():
  if key in result:
    result[key]+=value
  else:
    result[key]=value

# 3. Sắp xếp kết quả (Khắc phục lỗi thiếu requirement)
# sorted(result.items()) trả về list các cặp tuple đã sắp xếp theo key
# dict(...) chuyển lại thành dictionary để in cho đẹp
sorted_result = dict(sorted(result.items()))

print(sorted_result)
```

```python
#w6a13
# Giả sử input: "a 1 b 2 c 3"
raw_input = input().split()

# Bước 1: Tạo dict ban đầu từ chuỗi input
# Logic: Nhảy cóc 2 bước (i, i+2) để lấy cặp
# Lưu ý: Đề bài không yêu cầu value phải là số, nên cứ để string
old_dict = {raw_input[i]: raw_input[i+1] for i in range(0, len(raw_input), 2)}

# Bước 2: Đảo ngược (Core logic)
# Thay vì viết vòng lặp for dài dòng, ta dùng Comprehension
new_dict = {v: k for k, v in old_dict.items()}

print(new_dict)
```

```python
#w6a15
# Input dòng 1: "math 95 science 88 english 92"
raw_data = input().split()
# Input dòng 2: "90"
k = int(input())

# Bước 1: Parse input thành dict (Nhớ ép kiểu int cho value)
scores = {raw_data[i]: int(raw_data[i+1]) for i in range(0, len(raw_data), 2)}

# Bước 2: Lọc dữ liệu (Filtering)
# Cấu trúc: {key: val for ... in ... if điều_kiện}
result = {key: val for key, val in scores.items() if val > k}

print(result)
```

```python
#w6a21
a=input().split()
result={}
for i in range(len(a)):
    # Sử dụng result.get() để lấy tuple hiện tại hoặc một tuple rỗng nếu khóa chưa tồn tại
    # Sau đó nối thêm chỉ số hiện tại vào tuple đó
    result[a[i]] = result.get(a[i], ()) + (i,)
print(result)
```
